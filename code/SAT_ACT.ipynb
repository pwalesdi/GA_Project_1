#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
import math
````

````python
def st_dev(list_x):
    x_minus_xbar_sq = 0
    for x in list_x:
        ssqs = ((x - np.mean(list_x))**2)
        x_minus_xbar_sq += ssqs
        ssqs_1 = math.sqrt((x_minus_xbar_sq) / (len(list_x)))
    return ssqs_1
````

````python
import seaborn as sns
plt.figure(figsize=(16, 9))

corr = act_sat1718.corr()
# Set up mask to be "True" in the upper triangle.
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
plt.title("ACT composite vs. SAT Composite 2017 & 2018", fontsize=15);


ax = sns.heatmap(corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(125, 175, n=200),
    mask = mask,
    square=True,
    annot=True)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
````
